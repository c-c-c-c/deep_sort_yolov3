```
# 標準使用ライブラリー
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import warnings
warnings.filterwarnings('ignore')
import gc
import os
import shutil
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

```
pi = float(np.pi)

vec_u = []

for tmp_a in np.random.rand(100):
    vec_u.append(10 * np.sin(tmp_a * 2*pi))
```

```
fig = plt.figure(figsize=(12, 8)) #...1

x = range(100)
y = [np.exp(xx) for xx in vec_u ]

# Figure内にAxesを追加()
ax = fig.add_subplot(111) #...2
ax.plot(x, y, label="test") #...3
```

```
fig = plt.figure(figsize=(12, 8)) #...1

x = range(100)

y = [np.exp(xx) for xx in vec_u ]
z = sum (y)

yy = y/z

# Figure内にAxesを追加()
ax = fig.add_subplot(111) #...2
ax.plot(x, yy, label="test") #...3
```
